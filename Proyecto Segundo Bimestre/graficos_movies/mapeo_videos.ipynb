```python
import numpy as np
import pandas as pd
```

```python
path = "C://Users//User//Documents//GitHub//py-o-a-hinostroza-carol-lizeth//Proyecto Segundo Bimestre//movies_spider//movies.csv"
columnas = ['Titulo','Ranking','Duracion', 'Director','Recaudacion','Year']
moviesdf = pd.read_csv(path,encoding='latin-1',names=columnas)
moviesdf
```

```python
recaudacion_serie_str = moviesdf['Recaudacion'].values
recaudacion_serie_str
```

```python
def limpiar_monto (montostr):
    montostr= montostr.replace(',','')
    return int(montostr)
```

```python
limpiar_monto('3,435,047')
```

```python
recaudacion_array = list(map(limpiar_monto,recaudacion_serie_str))
```

```python
moviesdf.drop(['Recaudacion'],axis=1)
```

```python
moviesdf['Recaudacion']= recaudacion_array
```

```python
moviesdf
```

```python
year_str = moviesdf['Year'].values
year_str 
```

```python
def limpiar_year (yearstr):
    yearstr= yearstr.strip('()')
    return int(yearstr)
```

```python
limpiar_year('(201864564)')
```

```python
year_array = list(map(limpiar_year,year_str))
```

```python
moviesdf.drop(['Year'],axis=1)
```

```python
moviesdf['Year']= year_array
```

```python
moviesdf
```

```python
tiempo_str = moviesdf['Duracion'].values
tiempo_str
```

```python
def limpiar_tiempo (tiempostr):
    tiempostr= tiempostr.replace('min','')
    return int(tiempostr)
```

```python
limpiar_tiempo('5 min')
```

```python
tiempo_array = list(map(limpiar_tiempo,tiempo_str))
```

```python
moviesdf.drop(['Duracion'],axis=1)
```

```python
moviesdf['Duracion']= tiempo_array
```

```python
moviesdf
```

```python
import pandas as pd
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "C://Users//User//Documents//GitHub//py-o-a-hinostroza-carol-lizeth//Proyecto Segundo Bimestre//movies_spider//movies_proyecto.csv"
columnas = ['Titulo', 'Ranking','Duracion', 'Recaudacion','Director', 'Year']
movies_df= pd.read_csv(path, usecols = columnas)
movies_df
```

## 1. Películas con mayor Rating

```python
df_rating = movies_df.sort_values(by=['Ranking'],ascending=False).head(5)
rating_serie = pd.Series(df_rating['Ranking'].values, index=df_rating['Titulo'].values)
y_pos = np.arange(len(rating_serie))
plt.barh(y_pos, rating_serie.values, align='center', alpha=0.5)
plt.yticks(y_pos, rating_serie.index)
for i, v in enumerate(rating_serie.values):
    plt.text(v-3, i , str(v), color='black', fontweight='bold')
plt.xlabel('Rating')
plt.title('Películas con mayor Rating')
plt.show()
```

## 2. Películas con mayor Rating por año

```python
df_rating_2018 = movies_df.loc[movies_df['Year'] == 2018].sort_values(by=['Ranking'],ascending=False).head(5)
df_rating_2019 = movies_df.loc[movies_df['Year'] == 2019].sort_values(by=['Ranking'],ascending=False).head(5)
rating_serie_2018 = pd.Series(df_rating_2018['Ranking'].values, index=df_rating_2018['Titulo'].values)
rating_serie_2019 = pd.Series(df_rating_2019['Ranking'].values, index=df_rating_2019['Titulo'].values)
y_pos = np.arange(len(df_rating_2018))
fig, axs = plt.subplots(2)
#Figura 1
axs[0].barh(y_pos, rating_serie_2018.values, align='center', alpha=0.5)
axs[0].set_yticklabels(rating_serie_2018.index, fontdict=None, minor=False)
for i, v in enumerate(rating_serie_2018.values):
    axs[0].text(v-3, i , str(v), color='black', fontweight='bold')

#Figura2
axs[1].barh(y_pos, rating_serie_2019.values, align='center', alpha=0.5)
axs[1].set_yticklabels(rating_serie_2019.index, fontdict=None, minor=False)
for i, v in enumerate(rating_serie_2019.values):
    axs[1].text(v-3, i , str(v), color='black', fontweight='bold')
plt.show()
```

## 4. Películas con mayores ingresos por año

```python
df_ingresos_2018 = movies_df.loc[movies_df['Year'] == 2018].sort_values(by=['Recaudacion'],ascending=False).head(5)
df_ingresos_2019 = movies_df.loc[movies_df['Year'] == 2019].sort_values(by=['Recaudacion'],ascending=False).head(5)
ingresos_serie_2018 = pd.Series(df_ingresos_2018['Recaudacion'].values, index=df_ingresos_2018['Titulo'].values)
ingresos_serie_2019 = pd.Series(df_ingresos_2019['Recaudacion'].values, index=df_ingresos_2019['Titulo'].values)
y_pos = np.arange(len(df_ingresos_2018))
fig, axs = plt.subplots(2)
#Figura 1
axs[0].barh(y_pos, ingresos_serie_2018.values, align='center', alpha=0.5)
axs[0].set_yticklabels(ingresos_serie_2018.index, fontdict=None, minor=False)
for i, v in enumerate(ingresos_serie_2018.values):
    axs[0].text(v-3, i , str(v), color='black', fontweight='bold')
#Figura2
axs[1].barh(y_pos, ingresos_serie_2019.values, align='center', alpha=0.5)
axs[1].set_yticklabels(ingresos_serie_2019.index, fontdict=None, minor=False)
for i, v in enumerate(ingresos_serie_2019.values):
    axs[1].text(v-3, i , str(v), color='black', fontweight='bold')
plt.show()
```

## 5. Suma de ingresos por director

```python
serie_ingresos_director =movies_df.groupby('Director').sum()['Recaudacion'].sort_values(ascending=False).head(5)
y_pos = np.arange(len(serie_ingresos_director))
plt.barh(y_pos, serie_ingresos_director.values, align='center', alpha=0.5)
plt.yticks(y_pos, serie_ingresos_director.index)
for i, v in enumerate(serie_ingresos_director.values):
    plt.text(v-3, i , str(v), color='black', fontweight='bold')
plt.xlabel('Recaudacion')
plt.title('Directores con mayor recaudación')
plt.show()
```

## 6. Promedio de ratings por director

```python
serie_director_promedio =movies_df.groupby('Director').mean()['Ranking'].sort_values(ascending=False).head(5)
y_pos = np.arange(len(serie_director_promedio))
plt.barh(y_pos, serie_director_promedio.values, align='center', alpha=0.5)
plt.yticks(y_pos, serie_director_promedio.index)
for i, v in enumerate(serie_director_promedio.values):
    plt.text(v-3, i , str(v), color='black', fontweight='bold')
plt.xlabel('Rating')
plt.title('Directores con mayor promedio de rating')
plt.show()
```

## 7. Directores con más películas

```python
serie_director_peliculas =movies_df['Director'].value_counts().head(5)
y_pos = np.arange(len(serie_director_peliculas))
plt.barh(y_pos, serie_director_peliculas.values, align='center', alpha=0.5)
plt.yticks(y_pos, serie_director_peliculas.index)
for i, v in enumerate(serie_director_peliculas.values):
    plt.text(v, i , str(v), color='black', fontweight='bold')
plt.xlabel('Recaudacion')
plt.title('Directores con mayor recaudación')
plt.show()
```

## 8. Peliculas con mayor duración

```python
df_duracion = movies_df.sort_values(by=['Duracion'],ascending=False).head(5)
duracion_serie = pd.Series(df_duracion['Duracion'].values, index=df_duracion['Titulo'].values)
y_pos = np.arange(len(duracion_serie))
plt.barh(y_pos, duracion_serie.values, align='center', alpha=0.5)
plt.yticks(y_pos, duracion_serie.index)
for i, v in enumerate(duracion_serie.values):
    plt.text(v-3, i , str(v), color='black', fontweight='bold')
plt.xlabel('Duración (mins)')
plt.title('Películas con mayor duración')
plt.show()
```

## 9. Total de ingresos por año

```python
serie_ingresos_anio =movies_df.groupby('Year').sum()['Recaudacion'].sort_values(ascending=False).head(5)
y_pos = np.arange(len(serie_ingresos_anio))
plt.barh(y_pos, serie_ingresos_anio.values, align='center', alpha=0.5)
plt.yticks(y_pos, serie_ingresos_anio.index)
for i, v in enumerate(serie_ingresos_anio.values):
    plt.text(v-3, i , str(v), color='black', fontweight='bold')
plt.xlabel('Recaudacion')
plt.title('Años con mayor recaudación')
plt.show()
```

## 10. Relación Rating vs Recaudación

```python
df_10=movies_df.loc[movies_df['Recaudacion'] >0]
serie_rating_ingresos = pd.Series(df_10['Recaudacion'].values,index=df_10['Ranking'].values).groupby(level=0).sum()
fig, ax = plt.subplots()
ax.plot(serie_rating_ingresos.index, serie_rating_ingresos.values)

ax.set(xlabel='Rating', ylabel='Recaudación',
       title='Relación entre Rating y Recaudación')
ax.grid()

z = np.polyfit(serie_rating_ingresos.index, serie_rating_ingresos.values, 1)
p = np.poly1d(z)
plt.plot(serie_rating_ingresos.index,p(serie_rating_ingresos.index),"r--")
plt.show()
```

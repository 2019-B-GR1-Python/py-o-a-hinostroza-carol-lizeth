```python
import numpy as np
from scipy import ndimage
from scipy import misc
```

```python
mapache = misc.face()
mapache
```

```python
mapache.shape  # SHAPE para ver las dimensiones del arreglo
```

```python
type(mapache)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(mapache)
```

```python
plt.imshow(mapache[::-1])
```

```python
plt.imshow(mapache[::2])
```

```python
plt.imshow(mapache[::-2])
```

```python
oscurito = ndimage.uniform_filter(mapache, 5)
plt.imshow(oscurito)
```

```python
local_mean = ndimage.uniform_filter(mapache, size=20)
plt.imshow(local_mean)
```

```python
blurred_l = ndimage.gaussian_filter(mapache, 10)
plt.imshow(blurred_l)
```

```python
filter_blurred_l = ndimage.gaussian_filter(blurred_l, 1)
alpha = 30
sharpened = blurred_l + alpha * (blurred_l - filter_blurred_l)
plt.imshow(sharpened)
```

```python
rotado = ndimage.rotate(mapache, 45,reshape=False)
plt.imshow(rotado)
```
